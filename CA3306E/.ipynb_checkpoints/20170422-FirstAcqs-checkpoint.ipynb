## Testing the acquisition.

```python
File = "data/1MHz.data"
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open(File) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
V = [x.strip().split("\t")[1] for x in content] 
V = np.array(V).astype(np.float)*2.0/3.0
t = [x.strip().split("\t")[0] for x in content]
t = np.array(t).astype(np.float)/1000.0
```

```python
RefT = 8000
plt.plot(t[RefT+0:RefT+50],V[RefT+0:RefT+50])  # plotting 50x 100ns, that's 5Âµs
plt.ylabel('ADC output')
plt.xlabel("Time (us)")
plt.show()
```
